```python
# Korelasyonu yÃ¼ksek olan ikilileri bul
korelasyon_threshold = 0.8
korelasyonlu_ikililer = []

for sutun in korelasyon_matrisi.columns:
    korelasyonlu_sutunlar = korelasyon_matrisi[sutun][abs(korelasyon_matrisi[sutun]) > korelasyon_threshold].index.tolist()
    korelasyonlu_ikililer.extend([(sutun, k) for k in korelasyonlu_sutunlar if k != sutun])

# SonuÃ§larÄ± bastÄ±r
for ikili in korelasyonlu_ikililer:
    print(f"{ikili[0]} ve {ikili[1]} arasÄ±nda yÃ¼ksek korelasyon bulundu.")
```

```python
df_copy=df.copy()
```

```python
df_copy["VAR_1"].mode()[0]
```

```python
def pooling_mode(df):
    for i in df.keys():
        df[i]=df[i].fillna(df[i].mode()[0],inplace=False)
    return df
        
```

```python
df_copy=pooling_mode(df_copy)
```
